```python
import requests
import os
import pandas as pd
import numpy as np
```

```python
url = 'https://swapi.dev/api/people/'
people = requests.get(url).json()
people
```

```python
people_ = []
for i in range(1, people['count'] + 1):
    content = requests.get(os.path.join(url, str(i)))
    if content.status_code != 404:
        people_.append(content.json())
```

```python
# birth year
birth_year = pd.Series([i['birth_year'] for i in people_])

# the oldest age
age_oldest = np.max(birth_year[birth_year.str.endswith('BBY')].str.replace('[a-zA-Z]+', '').astype(float))

# location the oldest age
index = int(np.where(birth_year == str(int(age_oldest)) + 'BBY')[0])

print('Name of the oldest creature:', people_[index]['name'])
```

```python
film = people_[index]['films']
titles = []
release_dates = []
for i in range(len(film)):
    film_ = requests.get(film[i]).json()
    titles.append(film_['title'])
    release_dates.append(film_['release_date'])
    
df_title = pd.DataFrame(dict(title = titles, release_date = release_dates))
df_title
```
