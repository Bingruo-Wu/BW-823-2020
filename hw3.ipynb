```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url1 = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2018/2018-11-13/malaria_inc.csv'
df_inc = pd.read_csv(url1,index_col=0,parse_dates=[0])
```

```python
dfcn = df_inc[df_inc['Code'] == 'CHN']
plt.plot(dfcn['Year'],dfcn['Incidence of malaria (per 1,000 population at risk) (per 1,000 population at risk)'])
plt.title('Incidence of Malaria/1000 population at risk, China')
plt.ylabel('Incidence')
plt.xlabel('Year')
plt.show()
```

```python
url2 = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2018/2018-11-13/malaria_deaths_age.csv'
df_age = pd.read_csv(url2,index_col=0,parse_dates=[0])
```

```python
url3 = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2018/2018-11-13/malaria_deaths.csv'
df_death = pd.read_csv(url3,index_col=0,parse_dates=[0])
```

```python
dfafg = df_death[df_death['Code'] == 'AFG']
plt.plot(dfafg['Year'],dfafg['Deaths - Malaria - Sex: Both - Age: Age-standardized (Rate) (per 100,000 people)'])
plt.title('Age of death from Malaria/1000 people, Afghanistan')
plt.ylabel('Age')
plt.xlabel('Year')
plt.show()
```

```python
ns = []
for i in df_age.year.unique():
    ns.append(df_age.loc[df_age['year'] == i, 'deaths'].sum())
d = {'Year':df_age.year.unique(),'Deaths':np.log(ns)}
dd = pd.DataFrame(d, columns=['Year','Deaths'])
```

```python
plt.plot(dd['Year'],dd['Deaths'])
plt.title('Log of the Number of Deaths from Malaria, globally')
plt.ylabel('log(number)')
plt.xlabel('Year')
plt.show()
```
